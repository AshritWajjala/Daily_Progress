# **Chi Square Test**

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import pickle
import seaborn as sns
import scipy.stats as stats
```

```python
df = sns.load_dataset('tips')
```

```python
df.head()
```

```python
dataset = pd.crosstab(df['sex'], df['smoker'])
```

```python
dataset.head()
```

```python
observed_values = dataset.values
```

```python
expected_values, dof = stats.chi2_contingency(dataset)[3], stats.chi2_contingency(dataset)[2]
```

```python
dof
```

```python
expected_values
```

```python
p_value = stats.chi2_contingency(dataset)[1]
p_value
```

```python
chi_square = sum([((o-e)**2)/e for o, e in zip(observed_values, expected_values)])
chi_square = chi_square[0] + chi_square[1]
print(f"chi2_statistic: {chi_square}")
```

```python
alpha = 0.05
critical = stats.chi2.ppf(q=1 - alpha, df=dof)
critical
```

```python
p_value = 1 - stats.chi2.cdf(x=chi_square, df=dof)
p_value
```

```python
if chi_square >= critical:
    print("Reject H0, There is a relationship between 2 categorical variables")
else:
    print("Retain H0, There is no relationship between 2 categorical variables")
```
