# **One sample T test**
Used when you want to test whether its mean is equal to a specific value

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import pickle
import seaborn as sns
import scipy.stats as stats
```

```python
ages=[10,20,35,50,28,40,55,18,16,55,30,25,43,18,30,28,14,24,16,17,32,35,26,27,65,18,43,23,21,20,19,70]
```

```python
import random

sample = random.sample(ages, 10)
```

```python
sample_mean = np.array(sample).mean()
pop_mean = np.array(ages).mean()
pop_mean
```

```python
# Computing t-test statistic and p_value using in-build method

ttest, p_value = stats.ttest_1samp(sample, pop_mean)
```

```python
print(ttest)
print(p_value)
```

```python
if p_value < 0.05:
  print("Reject nul hypothesis")
print("Retain null hypothesis")
```

# **Another example using poisson distribution**

```python
pop = stats.poisson.rvs(loc=20, mu=30, size = 1000)
```

```python
sample = stats.poisson.rvs(loc=20, mu=35, size=60)
```

```python
pop_mean = pop.mean()
```

```python
ttest, p_value = stats.ttest_1samp(sample, pop_mean)
```

```python
print(f"t-test statistic: {ttest}")
print(f"p-value: {p_value}")
```

```python
print("Reject Null Hypothesis" if p_value < 0.05 else "Retain Null Hypothesis")
```

# **Two-Sampled t-test**
Used to compare means of different groups (Null hypothesis --> equal means)

```python
classA_ages = stats.poisson.rvs(loc=20, mu=35, size = 50)
classB_ages = stats.poisson.rvs(loc=20, mu=30, size = 50)
```

```python
ttest_statistic, p_value = stats.ttest_ind(classA_ages, classB_ages, equal_var=False)
p_value
```

```python
print("Reject Null Hypothesis" if p_value < 0.05 else "Retain Null Hypothesis")
```

# **Paired t-test**
Used to check if there is any significant difference in the groups.

```python
weight_before = [25,30,28,35,28,34,26,29,30,26,28,32,31,30,45]
# Added noise for demonstration
weight_after = weight_before + stats.norm.rvs(scale=5, loc=-1, size=len(weight_before))
```

```python
weight_df = pd.DataFrame({"Weight Before" : np.array(weight_before),
                         "Weights After": np.array(weight_after),
                         "Difference": np.array(weight_before - weight_after)})
```

```python
weight_df
```

```python
ttest_statistic, p_value = stats.ttest_rel(weight_before, weight_after)
p_value
```

```python
print("Reject Null Hypothesis" if p_value < 0.05 else "Retain Null Hypothesis")
```

# **Exploratory Data Analysis (EDA)**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

# Display all the columns of the dataframe
pd.pandas.set_option('display.max_columns',None)
```

```python
dataset = pd.read_csv('train.csv')
```

```python
dataset.head()
```

```python
# Print the shape of the dataset
print(dataset.shape)
```

## **In the Data Analysis, we will analyze to find all the below**
1. Missing values
2. All the numerical variables
3. Distribution of numerical variables
4. Categorical variables.
5. Outliers
6. Relationship between dependent and independent variables.
7. Correlations

```python
# Missing values
dataset.isnull()
```

```python
# Sum of missing values
dataset.isnull().sum()
```

```python
# Returning the list of features which are having missing values (NaN)
nan_features = [feature for feature in dataset.columns if dataset[feature].isnull().sum()>=1]
nan_features
```

```python
# Returning the percentage of missing (NaN) values per feature
for feature in nan_features:
  print(f"{feature}: {np.round(dataset[feature].isnull().mean()*100, 4)}% missing values")
```

```python
sns.boxplot(dataset['SalePrice'])
```

As we can observe, there are so many outliers in the sales price column. So, we can use median instead of mean.

```python
# Plotting some diagrams
data = dataset.copy()

for feature in nan_features:
  data[feature] = np.where(data[feature].isnull(), 1, 0)

  data.groupby(feature)['SalePrice'].median().plot.bar()
  plt.title(feature)
  plt.show()
```

The relationship between missing values and deoendent variable (sales price) is shown above, which has to be handled meaningfully using feature engineering.

### Numerical *Features*

```python
# Checking whether the sales price column is object dtype
dataset['SalePrice'].dtypes!='object'
```

```python
# Numerical features
numerical_features = [feature for feature in dataset.columns if dataset[feature].dtypes!='object']
dataset[numerical_features].head()
```

```python
# Temporal variables --> Variables that are assoiciated with time. Eg., DateTime variables
year_features = [feature for feature in numerical_features if 'Yr' in feature or 'Year' in feature]
year_features
```

```python
for feature in year_features:
  print(feature, dataset[feature].unique())
```

```python
dataset.groupby('YrSold')['SalePrice'].median().plot()
plt.xlabel('Year Sold')
plt.ylabel('Median House Price')
plt.title('House Price vs Year Sold')
plt.show()
```

Now, we will plot a scatter plot that illustrates the age of the house v/s the sale price

```python
for feature in year_features:
  if feature != 'YrSold':
    data = dataset.copy()
    data[feature] = data['YrSold'] - data[feature]
    plt.scatter(data[feature], data['SalePrice'])
    plt.title(f"{feature} v/s Sales Price")
    plt.xlabel(feature)
    plt.ylabel("Sales Price")
    plt.show()
```

## Obervations:
1. Follows power law distribution (pareto distribution).
2. The sales price was decreasing with time (the cost of new house is higher than old house).


```python
# Returning the number of Discrete (numerical) features present in the dataset

discrete_features = [feature for feature in numerical_features if len(dataset[feature].unique()) <= 25]
print(f"Total number of discrete features: {len(discrete_features)}")
```

```python
# Display relation between Sale Price and discrete features

for feature in discrete_features:
  data = dataset.copy()
  data.fillna({feature: data[feature].median()}, inplace=True)
  plt.figure(figsize=(8, 5))
  data.groupby(feature)['SalePrice'].median().plot.bar()
  plt.xlabel(feature)
  plt.ylabel('Median House Price')
  plt.title(f"{feature} v/s Median Sale Price")
  plt.show()
```

### From the above plots, we can say that there is relationship between discrete variables and sales price.

```python
# Extracting continuous variables
continuous_features = [feature for feature in numerical_features if feature not in discrete_features+year_features]
print(len(continuous_features))
```

### Analyzing the continuous variables by plotting histograms for understanding the distribution

```python
data = dataset.copy()

for feature in continuous_features:
  sns.histplot(data[feature], kde=True)
  plt.xlabel(feature)
  plt.ylabel("Count")
  plt.title(feature)
  plt.show()
```

```python
# As we can see that most of the distributions are log normal, we will transform those to normal by using log transformation
data = dataset.copy()
for feature in continuous_features:
  if 0 in data[feature].unique():
    pass
  else:
    data[feature] = np.log(data[feature])
    # Now plot Scatter plot
    plt.scatter(data[feature], data['SalePrice'])
    plt.xlabel(feature)
    plt.ylabel('SalePrice')
    plt.title(f"{feature} v/s Sale Price")
    plt.show()

```

```python
# For Analyzing outliers in each continuous variable, plotting box plots

data = dataset.copy()
for feature in continuous_features:
  if 0 in data[feature].unique():
    pass
  else:
    data[feature] = np.log(data[feature])
    # Now plot Box plot
    data.boxplot(column=feature)
    # plt.xlabel(feature)
    # plt.ylabel('SalePrice')
    plt.title(feature)
    plt.show()

```

```python
# Displaying the change in distribution of kde of Sale Price before and after log transformation

plt.figure(figsize=(14, 6))
plt.subplot(1, 2, 1)
sns.kdeplot(dataset['SalePrice'])
plt.title("Before Log Transfromation")

plt.subplot(1, 2, 2)
sns.kdeplot(np.log(dataset['SalePrice']))
plt.title("After Log Transfromation")
```
